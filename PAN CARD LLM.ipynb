```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter

# Code to define document_list from cell 0d9T-71fSlq0
loader = TextLoader("/content/dataset.txt")
document_list = loader.load()

text_splitter = CharacterTextSplitter(
    chunk_size=1000, # Increased chunk size
    chunk_overlap=100 # Increased chunk overlap
)

docs = text_splitter.split_documents(document_list)

len(docs)
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
```

```python
from langchain_community.vectorstores import FAISS

db = FAISS.from_documents(docs, embeddings)
```

```python
query = "Who needs a PAN card?"

results = db.similarity_search(query, k=2)

for r in results:
    print(r.page_content)
    print("------")
```
